```python
#Input -  YearsExperience 
#Output - Salary
#divide the data into input(x) and output(y)
x = df.iloc[:,0:1].values
x
```

```python
y = df.iloc[:,1].values
y
```

```python
#We are not performing step no 5 and step no 6 for this dataset(limited data)
```

```python
#7.Apply a CLASSIFIER/REGESSOR/CLUSTER
#sklearn.linear_model - Package(collection),LinearRegression - Library Name
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#8.fit the model(Mapping of inputs with outputs)
#LinearRegression.fit
model.fit(x,y) #we are plotting values of x and y inside the Linear Regression library
```

```python
#9.predict the output
y_pred = model.predict(x)
y_pred
```

```python
y #actual output
```

```python
#Conclusion
#we have to compare y and y_pred
#This only means that my model is NOT linear
```

```python
#Individual Prediction
model.predict([[39891.0]])
```

```python
#Cross Verification Technique
#y=mx+c #equation of a staright line
m = model.coef_ #slope
m
```

```python
c = model.intercept_ #Y - intercept/constant
c
```

```python
#y = mx + c , x=39891.0
m * 39891.0 + c
```

```python
#Visulisation - Best Fit Line
plt.scatter(x,y)
plt.plot(x,y_pred, c = 'red')
plt.title('BEST FIT LINE')
```
